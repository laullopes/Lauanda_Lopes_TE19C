```python
import pandas as pd 

# Uppgift 2 - a)

df = pd.read_csv("sysselsattning.csv", encoding="ISO-8859-1", header=1)

df_män = df[df["kön"] == "män"]

ja = df_män.head(6)
total_m = ja["2019"].sum()
print(f"Antalet män som förvärvsarbetade: {total_m}")


df_kvinnor = df[df["kön"] == "kvinnor"]
yes = df_kvinnor.head(6)
total_k = yes["2019"].sum()
print(f"Antalet kvinnor som förvärvsarbetade: {total_k}" )
```
