# Miljöutsläpp 

Ett miljöutsläpp halverar fågelpopulation i en ö varje år.

Fågelpopulationen började på 8000 st 

Hur många år det tills populationen blir mindre än en tiondel?

```python
fåglar = 8000
antal_år = 0 

print( f"år{antal_år}: antal fåglar:{fåglar}st")

while fåglar >= 800:
    fåglar /= 2
    antal_år += 1
    print( f"år{antal_år}: antal fåglar:{fåglar}st") 

print(f"Det tar {antal_år} år tills fågelpopulationen understiger en tiondel")
```

```python
'''
1L mjölk har 1 500 000 bakterier
- I rumstemp ökar antalet bakterier med 50% varje timme 
- Mjölk blir sur när bakterierna överstiger 10 000 000 st 
- Hur många timmar tills mjölken blir sur?

'''
import matplotlib.pyplot as plt
import numpy as np #den här modelen är bra för matte, den hjälper med att skapa en lista 

bakterier = 1.5e6 # 1.5*10**6
surt = 1e7 
faktor = 1.5 
timmar = 0 

y_bakterier = [bakterier] #lista som ska hålla bakteriernas "förändring"

while bakterier < surt:
    bakterier = faktor*bakterier
    y_bakterier.append(bakterier) #lägger på bakterier till vår lista 
    timmar += 1
x_timmar = np.arange(timmar+1)


print(f"Det tar {timmar}h för att mjölken ska surna") 
print(f"y = {y_bakterier}")
print(f"x = {x_timmar}")

plt.plot(x_timmar, y_bakterier)
plt.xlabel("Timmar(h)")
plt.ylabel("Antal bakterier")
plt.title("Bakterietillväxt i mjölk i rumstemperatur")
plt.show()
```
