```python
import pandas as pd 
import matplotlib.pyplot as plt

# Uppgift 3 

df = pd.read_csv("sysselsattning.csv", encoding="ISO-8859-1", header=1) 
df_män = df[df["kön"] == "män"]
ålder = df["ålder"].unique()
förv_män = df_män.head(6)
antal = förv_män["2019"].tolist()



df_kvinnor = df[df["kön"] == "kvinnor"]
ålder = df["ålder"].unique()
förv_kvi = df_kvinnor.head(6)
antal_kvi = förv_kvi["2019"].tolist()


plt.plot(ålder, antal, color= "blue" , label='Män')
plt.plot(ålder, antal_kvi, color= "pink", label= 'Kvinnor')
plt.legend(['Män', 'Kvinnor'])
plt.title("Förändring för förvävsarbetande män och kvinnor")
plt.show()

```
