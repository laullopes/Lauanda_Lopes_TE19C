```python
def medelvärde(a,b,c,d):
    medel = (a+b+c+d)/4 # lokal variabel 
    return medel 

# print(medel) # Det här blir syntax error 
print(medelvärde(1,2,3,4))
print(medelvärde(10,20,30,40)) # Fördelen är att man kan använda koden flera gånger med olika värde 
```

```python
# Räkna ord 
def räkna_ord(text):
    ord = text.split()
    antal_ord = len(ord)
    return antal_ord 


citat = " I stand on the shoulders of giants"

print(f"Antal ord i citatet: {räkna_ord(citat)}")

```

# Uppgift från Matte NP 

```python
import numpy as np 
import matplotlib.pyplot as plt 
def N(t):
    return 11/(1+3.4*np.exp(-0.03*t))
# a) 

print(f"a) Det är {N(0)} på jorden år 1950") 

# b) 
år = 100 

t= np.arange(år)

plt.plot (t,N(t))
plt.ylabel("ANtal människor i miljarder")
plt.xlabel("År efter 1950")
plt.title("Model för population")
plt.grid( )
plt.show( ) 

print(f"Övre gräns:{N(100)}")
```
