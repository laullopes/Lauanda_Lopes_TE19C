```python
import numpy as np 
import matplotlib.pyplot as plt 

# Uppgift 3 

# f(x) = 3x+1  0 <= x <= 10
x = np.linspace(0,11)
# print(f"x = {x}")

def f(k,x,m):
    y = k*x + m 
    return y 
# print(f"y = {3*x+1}")

plt.plot(x,f(3,x,1))
plt.title("En rätt linje")
plt.xlabel("x")
plt.ylabel("y")
plt.show()


```

```python
import numpy as np 
import matplotlib.pyplot as plt 

# Uppgift 4 
# f(x)= x^2 - 1  -2 <= x <= 2 

x = np.linspace(-2,2)
def f(x,m):
    y = x**2 - m 
    return y

plt.plot(x, f(x,1))
plt.title("Graf för linjen f(x)=x^2-1 för -2<=x<=2")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

# Uppgift 5 

x = np.linspace(-0.04,6)
def f(x):
    y = -4*x**2+24*x+1
    return y 

störst = np.max(f(x))
index = np.argmax(f(x))

print(f"Högsta höjden är vid ca {störst} m")
plt.plot(x,f(x))
plt.plot(x[index], störst, "r*")
plt.xlabel("tid i sekunder")
plt.ylabel("höjden i meter")
plt.show()
```
