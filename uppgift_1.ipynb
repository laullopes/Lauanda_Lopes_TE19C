e) Räkna antalet punkter som hamnar inne i cirkeln och dividera med antalet simulerade punkter.

```python
import random as rnd 

inside = 0 # skapat den variabeln för att kunna räkna antalet punkter som hamnar inne i cirklen. 

for k in range(10): 
    k = (math.sqrt(rnd.uniform(-1,1)**2 + rnd.uniform(-1,1)**2)) # k betyder kordinaterna för både x och y-axeln. 
    
    if k <= 1: # om k är mindre än pytahgorassats av x och y-axeln så ligger punkten innanför. 
        print(f"{k: 0f}", "innanför")
        inside = inside + 1 # den här delen loopar antalet punkter som hamnar inne i cirkeln. 
    
    elif k >= 1:
        print(f"{k: 0f}","utanför")

print("Andelen av punkterna som ligger i circkeln blir:", inside/10) # Antalet innanför punkter dividerad med alla punkter. 
```

g) Använd matplotlib och rita punkterna som hamnar inne i cirkeln och punkterna
som hamnar utanför. Använd två olika färger för att särskilja dem åt.

```python
import matplotlib.pyplot as plt  
import random

# Här gjorde jag om koden för att kunna rita grafen. 

for i in range(5000): 
    x = random.uniform(-1,1)
    y = random.uniform(-1,1)
    a = math.sqrt(x**2+y**2) 
    if a >= 1:
        plt.plot(x, y, "g*") # De gröna punkter ligger utanför cirklen. 
    elif a <= 1:
        plt.plot(x, y, "b*") # De blåa punkter ligger innanför cirkeln. 

# Ju fler punkter man har desto tydligare blir bilden. 
```
