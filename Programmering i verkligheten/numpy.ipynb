```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.arange(1,10)
print(f"x = {x}")

# y = kx + m 

def f(k,x,m):
    y = k*x + m 
    return y 

# y = 2x+2
print(f(2,x,2))

# y = x+1 
print(f"y = {f(1,x,1)}")

plt.plot(x,f(1,x,1))
plt.title(" En rät linje")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# tärningssimulering

```python
import numpy as np 
import random as rnd

utfall = np.zeros(6, dtype=int)
antal_kast = 100000

for i in range(antal_kast):
    kast = rnd.randint(1,6)
    utfall[kast-1] += 1 

print(f"Tärningskast:{utfall}")
print(f"Sannolikhet:{utfall/antal_kast}")
```

```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.linspace(-2,5)
def f(x):
    return -x**2+3*x+1

störts = np.max(f(x))
index = np.argmax(f(x))
print(x[index])

print(störts)

plt.plot(x,f(x))
plt.plot(x[index], störts,"r*")
plt.show()
```
